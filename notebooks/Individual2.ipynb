# Найти площадь фигуры ограниченной линиями: $$y=4x-x^2, y=4-xx$$ 

### Необходимо сначала найти точки пересечения двух кривых, заданными уравнениями $$y=4x-x^2, y=4-x$$
### Для этого, решаю уравнение: $$4x-x^2 = 4-x$$
$$4x-x^2 = 4-x \Rightarrow $$
$$ \Rightarrow x^2 - 3x + 4 = 0 \Rightarrow $$
$$ \Rightarrow x = \frac{3 ± \sqrt{9 - 16}} {2}$$

### Так как значение под корнем отрицательно, то точки пересечения находятся за пределами области определения, и следовательно фигура, ограниченная этими линиями, не имеет точек пересечения.
### Чтобы найти площадь фигуры, ограниченной этими линиями,  можно воспользоваться интегралом от разности этих функций на интервале от 0 до 4:

###  $$S = \int_{0}^4 (4x-x^2 - (4-x)) dx$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Добавление функции представляющей уравнение(интеграл)
def f(x):
    return 4*x - x**2 - (4-x)
```

```python
# Вычисление площади фигуры с помощью интеграла
a, b = 0, 4
n = 1000
x = np.linspace(a, b, n+1)
dx = (b-a)/n
y = f(x)
S = dx/2 * (y[0] + y[-1] + 2*np.sum(y[1:-1]))
```

```python
# Построение графика и вывод площади фигуры
fig, ax = plt.subplots()
ax.plot(x, y)
ax.fill_between(x, y, where=(y>=0), interpolate=True, alpha=0.2)
ax.set_xlim([a, b])
ax.set_ylim([0, 4])
plt.show()

print(f'Площадь фигуры: {S:.2f}')
```
