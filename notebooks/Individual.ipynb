Рассчитать средний рост учеников в классе: (1.64, 1.86, 1.72, 1.95, 1.76, 1.65, 1.79, 1.82, 1.92).

```python
from matplotlib import pylab as plt
```

```python
stat = [1.64, 1.86, 1.72, 1.95, 1.76, 1.65, 1.79, 1.82, 1.92]
count = len(stat)
l = 0

for i, co in enumerate(stat):
        l = l + stat[i]
        
print("Средний рост в классе: ", l/count)
```
